```python
# 인게임 정보 보는 함수
def ingame(name):
    import requests
    import pandas as pd
    api_key = 'RGAPI-9523cecd-825c-49d9-875e-f57ad8381c1a'
    sohwan = "https://kr.api.riotgames.com/lol/summoner/v4/summoners/by-name/" + str(name) +'?api_key=' + api_key
    r = requests.get(sohwan)
    r.json()['accountId'] #소환사의 고유 id
    active_url = "https://kr.api.riotgames.com/lol/spectator/v4/active-games/by-summoner/" +r.json()['id']+'?api_key=' + api_key
    active_champion  = requests.get(active_url).json()
    tmp_lst = []
    for i in range(10):
        tmp = []
        tmp.append(active_champion['participants'][i]['teamId'])
        tmp.append(active_champion['participants'][i]['spell1Id'])
        tmp.append(active_champion['participants'][i]['spell2Id'])
        tmp.append(active_champion['participants'][i]['championId'])
        tmp.append(active_champion['participants'][i]['summonerName'])
        for j in range(9):
            tmp.append(active_champion['participants'][i]['perks']['perkIds'][j])
        tmp.append(active_champion['participants'][i]['perks']['perkStyle'])
        tmp.append(active_champion['participants'][i]['perks']['perkSubStyle'])
        tmp.append(active_champion['bannedChampions'][i]['championId'])
        tmp.append(active_champion['gameLength'])
        tmp_lst.append(tmp)
    df = pd.DataFrame(tmp_lst)
    df = df.rename(columns={0:'teamId',1:'spell1Id',2:'spell2Id',3:'id',4:'summonerName',\
                           5:'main_rune_1',6:'main_rune_2',7:'main_rune_3',8:'main_rune_4',\
                           9:'sub_rune_1',10:'sub_rune_2',11:'detail_rune_1',12:'detail_rune_2',\
                           13:'detail_rune_3',14:'main_id',15:'sub_rune',16:'ban_championId',17:'gameLength'})
    rune_list = pd.read_csv('rune_list.csv')
    champ = pd.read_csv('champ.csv')
    main_rune_0 = rune_list[['main_id','main_name_e']]
    main_rune_1 = rune_list[['main_rune_1','main_rune_1ename']]
    main_rune_2 = rune_list[['main_rune_2','main_rune_2ename']]
    main_rune_3 = rune_list[['main_rune_3','main_rune_3ename']]
    main_rune_4 = rune_list[['main_rune_4','main_rune_4ename']]
    sub_rune = rune_list[['sub_rune','sub_rune_ename']]
    sub_rune_1 = rune_list[['sub_rune_1','sub_rune_1ename']]
    sub_rune_2 = rune_list[['sub_rune_2','sub_rune_2ename']]
    detail_rune_1 = rune_list[['detail_rune_1','detail_rune_1ename']]
    detail_rune_2 = rune_list[['detail_rune_2','detail_rune_2ename']]
    detail_rune_3 = rune_list[['detail_rune_3','detail_rune_3ename']]
    df = df.merge(main_rune_0).merge(main_rune_1).merge(main_rune_2).merge(main_rune_3).\
            merge(main_rune_4).merge(sub_rune).merge(sub_rune_1).merge(sub_rune_2).\
            merge(detail_rune_1).merge(detail_rune_2).merge(detail_rune_3)
    champ = champ[['name_e','id','name']]
    df = df.merge(champ, on='id')
    df = df[['teamId', 'name', 'name_e', 'ban_championId', 'spell1Id', 'spell2Id', 'summonerName', 'main_name_e',\
             'main_rune_1ename', 'main_rune_2ename', 'main_rune_3ename', 'main_rune_4ename',\
             'sub_rune_ename', 'sub_rune_1ename', 'sub_rune_2ename', 'detail_rune_1ename',\
             'detail_rune_2ename', 'detail_rune_3ename', 'gameLength']].rename(columns={'ban_championId':'id'})
    df = df.merge(champ, on='id')
    df = df[['teamId', 'name_x', 'name_e_x', 'name_e_y', 'name_y', 'id', 'spell1Id', 'spell2Id', 'summonerName', 'main_name_e',\
             'main_rune_1ename', 'main_rune_2ename', 'main_rune_3ename', 'main_rune_4ename',\
             'sub_rune_ename', 'sub_rune_1ename', 'sub_rune_2ename', 'detail_rune_1ename',\
             'detail_rune_2ename', 'detail_rune_3ename', 'gameLength']].rename(columns={'name_e_y':'ban_name', 'name_y':'ban_name_kr'})
    df = df[['teamId','name_x','name_e_x','ban_name_kr','ban_name','spell1Id',\
        'spell2Id','summonerName', 'main_name_e','main_rune_1ename', 'main_rune_2ename',\
         'main_rune_3ename', 'main_rune_4ename','sub_rune_ename', 'sub_rune_1ename',\
         'sub_rune_2ename', 'detail_rune_1ename','detail_rune_2ename', 'detail_rune_3ename',\
         'gameLength']].rename(columns={'name_x':'championName','name_e_x':'championName_e',\
                                       'ban_name_kr':'ban_championName','ban_name':'ban_championName_e'})
    spell = pd.read_csv('spell.csv')
    spell = spell[['id','name','name_e']]
    spell1 = spell.rename(columns={'id':'spell1Id','name':'spell1_name',\
                             'name_e':'spell_name_e'})
    spell2 = spell.rename(columns={'id':'spell2Id','name':'spell2_name',\
                             'name_e':'spell2_name_e'})
    df = df.merge(spell1, on='spell1Id')
    df = df.merge(spell2, on='spell2Id')
    df = df[['teamId', 'summonerName', 'championName', 'championName_e', 'ban_championName',\
          'ban_championName_e', 'spell_name_e', 'spell1_name', 'spell2_name_e', 'spell2_name',\
          'main_name_e', 'main_rune_1ename', 'main_rune_2ename',\
          'main_rune_3ename','main_rune_4ename', 'sub_rune_ename', 'sub_rune_1ename', 'sub_rune_2ename',\
          'detail_rune_1ename', 'detail_rune_2ename', 'detail_rune_3ename', 'gameLength']]
    return df
```
